```python
import pandas as pd
```

```python
# Load pacs_clean.csv
pacs_file = ("pacs.csv")
pacs_df = pd.read_csv(pacs_file, low_memory= False)
pacs_df.head()
```

```python
# Check data count
pacs_df.count()
```

```python
# Create unique_key column
pacs_df['unique_key'] = pacs_df['cycle'].map(str) + pacs_df['cid'].map(str)
pacs_df.head()
```

```python
# Reorder columns
pacs_df = pacs_df[["id", "cycle", "fec_rec_no", "pac_id", "unique_key", "cid", "amount", "date", "real_code", "type", "di", "fec_cand_id"]]
pacs_df.head()
```

```python
pacs_df.info()
```

```python
pacs_df['contribution_date'] = pacs_df['date'].str.slice(0, 10)
pacs_df
```

```python
# Rerun
# Change date column to datetime datatype
pacs_df['contribution_date'] = pd.to_datetime(pacs_df['contribution_date'])
pacs_df.info()
```

```python
# Drop date column
pacs_date_df = pacs_df.drop('date', axis=1)

pacs_date_df.head()
```

```python
# Reorder columns
pacs_date_df = pacs_date_df[["id", "cycle", "fec_rec_no", "pac_id", "unique_key", "cid", "amount", "contribution_date", "real_code", "type", "di", "fec_cand_id"]]
pacs_date_df.head(15)
```

```python
# Check for Null Values in contribution_date
pacs_date_df["contribution_date"].isnull().sum()
```

```python
# Check for Null Values in cid
pacs_date_df["cid"].isnull().sum()
```

```python
# Drop rows where cid is null
pacs_clean_df = pacs_date_df.dropna(subset=['cid'])
```

```python
# Check for Null Values in cid
pacs_clean_df["cid"].isnull().sum()
```

```python
pacs_clean_df.count()
```

```python
# Check for duplicates in id column
duplicates = pacs_date_df.id.duplicated().sum()
print(duplicates)
```

```python
# Save new file
pacs_clean_df.to_csv("pacs_clean.csv", encoding='utf-8', index=False)
```
